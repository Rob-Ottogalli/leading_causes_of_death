```python
#copy cleaned up dataframe to run specific analysis
seasons_df = year_2014_df_2
```

```python
#Drop irrelavent columns
seasons_df = seasons_df.drop(columns=['education_2003_revision', 'month_of_death', 'sex', 'detail_age', 'age_recode_27', 'infant_age_recode_22', 'marital_status', 'day_of_week_of_death', '39_cause_recode'])
seasons_df
```

```python
#Respitory = KEEP(247, 248, 251, 252, 258, 260, 261, 262, 276, 278, 361, 362, 371, 13, 90, 95)
#use iloc for respitory disease codes 
df251 = seasons_df[seasons_df.iloc[:, 0] == 251]
df260 = seasons_df[seasons_df.iloc[:, 0] == 260]
df261 = seasons_df[seasons_df.iloc[:, 0] == 261]
df276 = seasons_df[seasons_df.iloc[:, 0] == 276]
df278 = seasons_df[seasons_df.iloc[:, 0] == 278]
df361 = seasons_df[seasons_df.iloc[:, 0] == 361]
df362 = seasons_df[seasons_df.iloc[:, 0] == 362]
df371 = seasons_df[seasons_df.iloc[:, 0] == 371]
df13 = seasons_df[seasons_df.iloc[:, 0] == 13]
df95 = seasons_df[seasons_df.iloc[:, 0] == 95]

#Dropped several disease codes because no records
#Concact into one datafram for respitory diseases
respitory_df = pd.concat([df251, df260, df261, df276, df278, df361, df362, df371, df13, df95])
respitory_df

#Export to CSV
respitory_df.to_csv(r'03_Cleaned_Data\respitorymortality.csv', index = False)
```

```python
#Circulatory = KEEP(197, 244, 246, 370)
#use iloc for circulatory disease codes 
df246 = seasons_df[seasons_df.iloc[:, 0] == 246]
df370 = seasons_df[seasons_df.iloc[:, 0] == 370]

#Dropped 197 & 244 because there were no records 
#Concact into one datafram for respitory diseases
circulatory_df = pd.concat([df246, df370])
circulatory_df

#Export to CSV
circulatory_df.to_csv(r'03_Cleaned_Data\circulatorymortality.csv', index = False)
```
